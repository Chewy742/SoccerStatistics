## Visualizations : Bar Charts, Lollipop Bar Charts, Scatter Plots/Crosshairs

This is a notebook for reproducing some simple charts for data visualization. Use this for the framework for your own work/more complex data. 

### Region: Bar Charts

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 

%matplotlib inline

table = pd.read_csv("epl20162017.csv")
table.head()
```

```python
#X handles a range from 1 to 20, Y is arranged by points

plt.bar(x=np.arange(1,21),height=table['P'])
```

```python
#Time to make the graph readable

plt.bar(x=np.arange(1,21),height=table['P'])

plt.title ("Premier League 2016/17")

#xticks - Give the x axis some labels across the tick marks.
#Argument one is the position for each label
#Argument two is the label values and the final one is to rotate our labels

plt.xticks(np.arange(1,21), table['Team'], rotation=90)

#Give the x and y axes a title
plt.xlabel("Team")
plt.ylabel("Points")
plt.show()
```

```python
#Create an array of equal length to our bars
#Each value is the hex code for the team's colours, in order of our chart -- taken from FCPython

teamColours = ['#034694','#001C58','#5CBFEB','#D00027',
              '#EF0107','#DA020E','#274488','#ED1A3B',
               '#000000','#091453','#60223B','#0053A0',
               '#E03A3E','#1B458F','#000000','#53162f',
               '#FBEE23','#EF6610','#C92520','#BA1F1A']

#Add a new argument, color, to our 'plt.bar()' method
#This argument passes our teamColours array
plt.bar(x=np.arange(1,21),height=table['P'],color = teamColours)

plt.title("Premier League 2016/17")
plt.xticks(np.arange(1,21), table['Team'], rotation=90)
plt.xlabel("Team")
plt.ylabel("Points")
plt.show()
```
